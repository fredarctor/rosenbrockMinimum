```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#The Boltzmann constant and the cross-section for inelastic collision
boltz = 1.380649*(10**(-23))
sigma = 2.1*(10**(-19))

#Calculates the mean free path given a temperature in celsius
def meanfreepath(temp):
    kelvin = temp + 273.15
    
    index = 9 - (3110/kelvin)
    pressure = 8.7*(10**(index))
    
    return (boltz*kelvin)/(pressure*sigma)


print(boltz)
print(sigma)
print(meanfreepath(180))

#Temperatures recorded at and an empty array to fill with the mean free paths
temps = [150, 160, 170, 180, 190, 200]
meanpaths = [0, 0, 0, 0, 0, 0]

#Fill the above array with the mean free paths
for i in range(0, 6):
    meanpaths[i] = meanfreepath(temps[i])

#Plot the mean free path against temperature in celsius for the discrete temperatures above
plt.plot(temps, meanpaths)
plt.grid()
```

```python
plt.rcParams['font.size'] = 15
#Create some empty arrays for the data to follow
meanpathscont = np.zeros(62)
tempscont = np.zeros(62)

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(1, 1, 1)

ax.set_xlabel('Temperature, $T$ / C')
ax.set_ylabel('Mean Free Path,  / m')

#Fills the temperate array with values which are then operated on for the mean free path values
for i in range(139, 201):
    tempscont[i-139] = i
    meanpathscont[i-139] = meanfreepath(i)*(10**6)

#Plots a large array of mean free paths against their respective temperatures
ax.plot(tempscont, meanpathscont)
ax.grid(which = 'major', alpha = 0.9)
```
